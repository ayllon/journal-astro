```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from astropy.table import Table
```

```python
t = Table.read('/tmp/sourcex/SourceXtractorPlusPlus/x86_64-fc34-gcc112-dbg/test_single_frame.py/output.fits')
```

```python
idx = np.random.choice(len(t), size=36)

fig, axes = plt.subplots(6, 6, figsize=(22,22))

for i, ax in enumerate(axes.ravel()):
    obj = t[idx[i]]
    ax.imshow(obj['vignet'])
    if obj['pixel_centroid_x'] <= 8 or obj['pixel_centroid_y'] <= 8:
        ax.set_title('!!')
    if obj['pixel_centroid_x'] >= 1536-8 or obj['pixel_centroid_y'] >= 1536-8:
        ax.set_title('??')

plt.show()
```

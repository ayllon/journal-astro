```python
%matplotlib inline
import matplotlib.pyplot as plt
from astropy.table import Table, join

plt.rcParams['figure.figsize'] = (15, 8)
```

```python
orig = Table.read('/home/aalvarez/Work/PHZ_Calibration/run_calibration_in_parallel.iterations.1.CalibrationPhosphoros/phz_cat_original.fits')
opti = Table.read('/home/aalvarez/Work/PHZ_Calibration/run_calibration_in_parallel.iterations.1.CalibrationPhosphoros/phz_cat.fits')

x = join(orig, opti, 'ID')
```

```python
fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(15, 10))

axes[0, 0].set_title('Z')
axes[0, 0].scatter(x['Z_1'], x['Z_2'])
axes[0, 0].plot([0, 2.5], [0, 2.5], color='red')

axes[0, 1].set_title('SED')
axes[0, 1].scatter(x['SED-Index_1'], x['SED-Index_2'])
axes[0, 1].plot([0, 30], [0, 30], color='red')

axes[1, 0].set_title('EBV')
axes[1, 0].scatter(x['E(B-V)_1'], x['E(B-V)_2'])
axes[1, 0].plot([0, 0.5], [0, 0.5], color='red')

axes[1, 1].set_title('Scale')
axes[1, 1].scatter(x['Scale_1'], x['Scale_2'])
axes[1, 1].plot([0, 3.5e11], [0, 3.5e11], color='red')

plt.tight_layout()
```

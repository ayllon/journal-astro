```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def expr(a, b, c, d, e, f):
    assert e + f <= 1
    return 2**a * 3**b * 5**c * 7**d * 11**e * 13**f
```

```python
np.log2(6079)
```

```python
values = []
maxima = 2**13

for a in range(14):
    for b in range(8):
        for c in range(6):
            for d in range(5):
                for (e, f) in [(0,0), (0, 1), (1,0)]:
                    v = expr(a, b, c, d, e, f)
                    if v > maxima:
                        break
                    values.append(v)
values = np.sort(values)
```

```python
xs = np.arange(0, maxima+1)
ys = np.zeros_like(xs)

for i, x in enumerate(xs):
    idx = np.searchsorted(values, x)
    ys[i] = values[idx]
```

```python
plt.figure(figsize=(20, 10))
plt.plot(xs, ys)
```

```python
with np.printoptions(threshold=np.inf):
    print(np.array_repr(ys))
```
